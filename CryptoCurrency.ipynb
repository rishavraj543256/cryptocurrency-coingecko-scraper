```python
final=pd.DataFrame()
for i in range(1,133):
    url="https://www.coingecko.com/?page={}".format(i)
    response=requests.get(url)
    data=soup(response.content,'html.parser')
    div_data=data.findAll('div',{'class':'coin-table table-responsive lg:tw-table'})
    tbody=div_data[0].tbody.findAll('tr')
    coin=[]
    price=[]
    h=[]
    h_24=[]
    d_7=[]
    v_24=[]
    mkt_cap=[]
    img=[]
    for td in tbody:
        title=td.findAll('div',{'class':'tw-flex-auto'})
        coin.append(title[0].span.text.strip())
        pr=td.findAll('div',{'class':"tw-flex-1"})
        price.append(pr[0].text.strip())
        hr=td.findAll('td',{'class':'td-change1h change1h stat-percent text-right col-market'})
        h.append(hr[0].text.strip())
        t_4=td.findAll('td',{'class':'td-change24h change24h stat-percent text-right col-market'})
        h_24.append(t_4[0].text.strip())
        day_7=td.findAll('td',{'class':'td-change7d change7d stat-percent text-right col-market'})
        d_7.append(day_7[0].text.strip())
        volume=td.findAll('td',{'class':"td-liquidity_score lit text-right col-market"})
        v_24.append(volume[0].text.strip())
        cap=td.findAll('td',{'class':'td-market_cap cap col-market cap-price text-right'})
        mkt_cap.append(cap[0].text.strip())
        img_link=td.findAll('td',{'class':"p-0 pl-2 text-center"})
        img.append(img_link[0].img['src'])
    dict={'Coin':coin,"Price":price,'1h':h,'24h':h_24,"7d":d_7,'24h Volume':v_24,'Mkt Cap':mkt_cap,"Graph":img}
    df=pd.DataFrame(dict)
    final=final.append(df)
```
